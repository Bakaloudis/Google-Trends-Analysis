```python
# Dimitrios Bakalis, A.M: 3033

# Data Analysis for covid-19 based on:
# Google Trends

import pandas as pd 

from pandas import Series, DataFrame
from datetime import datetime 

import matplotlib.pyplot as plt
import matplotlib.pyplot as gridspec
%matplotlib inline

import seaborn as sns
import numpy as np
from numpy import median

import scipy as sp
from scipy import stats 
from scipy.stats import kendalltau, pearsonr, spearmanr
from scipy.stats import ttest_ind
```

```python
def worldwide_results(df_masks,df_vaccines,df_lockdown,df_corona,df_netflix,df_disney,df_hbo):

    main_col_list = ['face masks: (Worldwide)', 'vaccine: (Worldwide)', 'Lockdown: (Worldwide)', 'Coronavirus: (Worldwide)']
    df_dict = { 'face masks: (Worldwide)' : df_masks, 'vaccine: (Worldwide)' : df_vaccines, 'Lockdown: (Worldwide)' : df_lockdown, 'Coronavirus: (Worldwide)' : df_corona }
    name_dict = { 'face masks: (Worldwide)' : "Masks", 'vaccine: (Worldwide)' : "Vaccine", 'Lockdown: (Worldwide)' : "Lockdown", 'Coronavirus: (Worldwide)' : "Coronavirus" }

    frames = []

    for col in main_col_list:

        df_dict[col]['Week'] = pd.to_datetime(df_dict[col]['Week'], format = '%Y-%m-%d')
        df_dict[col] = df_dict[col].rename(columns={col: 'Data'})

        df_att_col = []
        index = len(df_dict[col].index)

        for i in range(0,index):
            df_att_col.append(name_dict[col])

        df_dict[col].insert(2, 'Attribute', df_att_col , True)

        frames.append(df_dict[col])

    results = pd.concat(frames)

    fig = plt.figure(figsize=(25,18))
    fig.subplots_adjust(hspace=0.3, wspace=0.2)

    ax = fig.add_subplot(3, 1, 1)
    sns.lineplot( x = "Week", y = "Data", data = results, hue = "Attribute", ax = ax)
    plt.grid()
    plt.xlabel("Months")
    plt.ylabel("WorldWide Data")

    main_col_list = ['Netflix: (Worldwide)', 'Disney+: (Worldwide)', 'HBO: (Worldwide)' ]
    df_dict = { 'Netflix: (Worldwide)' : df_netflix, 'Disney+: (Worldwide)' : df_disney , 'HBO: (Worldwide)' : df_hbo }
    name_dict = { 'Netflix: (Worldwide)' : "Netflix", 'Disney+: (Worldwide)' : "Disney+", 'HBO: (Worldwide)' : "HBO" }

    frames = []

    for col in main_col_list:

        df_dict[col]['Week'] = pd.to_datetime(df_dict[col]['Week'], format = '%Y-%m-%d')
        df_dict[col] = df_dict[col].rename(columns={col: 'Data'})

        df_att_col = []
        index = len(df_dict[col].index)

        for i in range(0,index):
            df_att_col.append(name_dict[col])

        df_dict[col].insert(2, 'Attribute', df_att_col , True)

        frames.append(df_dict[col])

    results = pd.concat(frames)

    ax = fig.add_subplot(3, 1, 2)
    sns.lineplot( x = "Week", y = "Data", hue = "Attribute", data = results, ax = ax)
    plt.grid()
    plt.xlabel("Months")
    plt.ylabel("Worldwide Data")

    plt.show()
```

```python

df_masks = pd.read_csv('face-masks.csv', skiprows=2)
df_vaccines = pd.read_csv('vaccine.csv', skiprows=2)
df_lockdown = pd.read_csv('lockdown.csv', skiprows=2)
df_corona = pd.read_csv('coronavirus.csv', skiprows=2)

df_netflix = pd.read_csv('netflix.csv', skiprows=2)
df_disney = pd.read_csv('disney.csv', skiprows=2)
df_hbo = pd.read_csv('hbo.csv', skiprows=2)

worldwide_results(df_masks,df_vaccines,df_lockdown,df_corona,df_netflix,df_disney,df_hbo)
```
